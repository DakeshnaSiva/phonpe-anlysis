```python
import os
import pandas as pd
import json

dir = r'D:\dakeshna\Django Python\link\pulse\data\aggregated\transaction\country\india\state'


dk = []

for root, dirs, files in os.walk(dir):
    for file in files:
        if file.endswith('.json'):
            file_path = os.path.join(root, file)
            try:
                with open(file_path, 'r') as f:
                    json_data = json.load(f)
                df = pd.json_normalize(json_data)
                dk.append(df)
            except Exception as e:
                print(f"Error reading file {file_path}: {str(e)}")
combined_df = pd.concat(dk, ignore_index=True)
print(combined_df)
```

```python
import os
import pandas as pd
import json

dir = r'D:\dakeshna\Django Python\link\pulse\data\aggregated\transaction\country\india\state'
dk = []
for year_dir in os.listdir(dir):
    year = year_dir[:4] 
    for quarter_dir in os.listdir(os.path.join(dir, year_dir)):
        quarter = quarter_dir.split('_')[-1]  
    
        for filename in os.listdir(os.path.join(dir, year_dir, quarter_dir)):
            if filename.endswith('.json'):
            
                filepath = os.path.join(dir, year_dir, quarter_dir, filename)
                with open(filepath, 'r') as f:
                    json_data = json.load(f)
                transaction_data = json_data['data']['transactionData']
                
                for transaction in transaction_data:
                    transaction_name = transaction['name']
                    for payment_instrument in transaction['paymentInstruments']:
                        payment_type = payment_instrument['type']
                        payment_count = payment_instrument['count']
                        payment_amount = payment_instrument['amount']
                        
                        # Create a new row for the dataframe
                        row = {
                            'year': year,
                            'quarter': quarter,
                            'transaction_type': transaction_name,
                            'transaction_count': payment_count,
                            'transaction_amount': payment_amount
                        }
                        
                        
                        dk.append(row)

                        
                     

df = pd.DataFrame(dk)
```

```python
year_to_quarter = {
    '2018': '1',
    '2019': '2',
    '2020': '3',
    '2021': '4'
}

df['quarter'] = df['year'].apply(lambda x: year_to_quarter.get(x, ''))
```

```python
df = df.rename(columns={'year': 'state', 'quarter': 'year'})
```

```python
import streamlit as st
import pandas as pd
import plotly.express as px


df = pd.read_excel('D:\dakeshna\Django Python\phonepe.xlsx')

st.set_page_config(page_title='My Dashboard')
st.title('My Dashboard')


state = st.sidebar.selectbox('Select a state', df['state'].unique())

filtered_df = df[df['state'] == state]


fig = px.bar(filtered_df, x='transaction_type', y='transaction_count', color='year',
             title='Transaction Count by Transaction Type')
st.plotly_chart(fig)


fig = px.line(filtered_df, x='year', y='transaction_amount', title='Transaction Amount Over Time')
st.plotly_chart(fig)

fig = px.scatter(filtered_df, x='transaction_count', y='transaction_amount', color='year',
                 title='Transaction Count vs. Transaction Amount')
st.plotly_chart(fig)
```
